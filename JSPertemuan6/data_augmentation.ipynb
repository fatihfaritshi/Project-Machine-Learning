```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageEnhance, ImageFilter
import os

# Path gambar (gunakan raw string atau double backslash)
image_path = r"C:\Users\Lenovo\OneDrive\Documents\SEMESTER 6 (FATIH)\Machine Learning\JSPertemuan6\fatih.png"

# Cek apakah file ada sebelum dibuka
if not os.path.exists(image_path):
    raise FileNotFoundError(f"File '{image_path}' tidak ditemukan. Pastikan path sudah benar.")

# Buka gambar dan konversi ke RGB untuk menghindari mode yang tidak didukung
image = Image.open(image_path).convert("RGB")

# Augmentasi manual tanpa imgaug
# Flip horizontal
flipped_image = image.transpose(Image.FLIP_LEFT_RIGHT)

# Rotasi gambar dalam rentang -10 sampai 10 derajat
rotated_image = image.rotate(np.random.uniform(-10, 10))

# Blur Gaussian sederhana dengan ImageFilter (pastikan gambar dalam mode RGB)
blurred_image = image.filter(ImageFilter.GaussianBlur(radius=1.0))

# Tampilkan hasil augmentasi
plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 1)
plt.imshow(flipped_image)
plt.title('Flip Horizontal')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(rotated_image)
plt.title('Rotasi')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(blurred_image)
plt.title('Blur')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
from textblob import TextBlob
import random
import nltk
nltk.download('punkt')

# Contoh teks
text = "Ini adalah contoh kalimat untuk augmentasi teks."

# Objek TextBlob untuk teks
blob = TextBlob(text)

# Pertambahan kata
augmented_text = blob.words + ['baru', 'kata']

# Penggantian kata acak
for i in range(len(blob.words)):
    if random.random() < 0.3: # Probabilitas 30% untuk penggantian kata
        augmented_text[i] = 'kata_baru'

# Konversi kembali ke teks
augmented_text = ' '.join(augmented_text)
print("Teks Asli:", text)
print("Teks Setelah Augmentasi:", augmented_text)
```
